```python
import queue

def findNeigh(node_name):
    sub_df_link_s = df_link[(df_link['source'] == node_name) & 
                            ((df_link['relation'] == 'r_request_jump') | (df_link['relation'] == 'r_dns_a') | 
                            (df_link['relation'] == 'r_subdomain') | (df_link['relation'] == 'r_cert'))]
    sub_df_link_t = df_link[(df_link['target'] == node_name) & 
                            ((df_link['relation'] == 'r_request_jump') | (df_link['relation'] == 'r_dns_a') | 
                            (df_link['relation'] == 'r_subdomain') | (df_link['relation'] == 'r_cert'))]
    tar = list(sub_df_link_s['target'])
    src = list(sub_df_link_t['source'])
    return tar + src

def bfs(node_name):
    queue_ = queue.Queue()
    visited = {}
    init = (node_name, None)
    queue_.put(init)
    frontier = set()
    while(not queue_.empty()):
        node_to_expand, pre = queue_.get()
        visited[node_to_expand] = pre
        if len(frontier) != 0:
            frontier.remove(node_to_expand)
        print('found:', len(visited), 'type:', node_to_expand[0])
        if len(visited) == 100: break
        neighbors = findNeigh(node_to_expand)
        for n in neighbors:
            if n not in visited.keys() and n not in frontier:
                queue_.put((n, node_to_expand))
                frontier.add(n)
        print(queue_.qsize())
    return visited

v = bfs('Domain_f3554b666038baffa5814c319d3053ee2c2eb30d31d0ef509a1a463386b69845')
```

```python
week_relations = ['r_cert_chain', 'r_cname', 'r_asn', 'r_cidr']
normal_relations = ['r_whois_name', 'r_whois_email', 'r_whois_phone']
strong_relations = ['r_cert', 'r_subdomain', 'r_request_jump', 'r_dns_a']

def ComputeNeighbor(node, relations):
    mask = (df_link['source'] == node) | (df_link['target'] == node)
    temp_mask = (df_link['relation'] == None)
    for rela in relations:
        temp_mask |= (df_link['relation'] == rela)
    mask &= temp_mask
    return mask
```

```python
def ComputeSubgraph(cruls):
    mask = (df_link['relation'] == None)
    for crul in cruls:
        # one time
        mask |= ComputeNeighbor(crul, week_relations)
        # two times
        mask1 = ComputeNeighbor(crul, normal_relations)
        normal_neighbor1 = df_link[mask1]
        neis = list(set(normal_neighbor1['source'])) + list(set(normal_neighbor1['target']))
        neis.remove(crul)
        mask2 = (df_link['source'] == None)
        for nei in neis:
            mask2 |= ComputeNeighbor(nei, strong_relations)
        mask |= (mask1 | mask2)
        # three times
        mask1 = ComputeNeighbor(crul, strong_relations)
        imp_neighbor1 = df_link[mask1]
        neis1 = list(set(imp_neighbor1['source'])) + list(set(imp_neighbor1['target']))
        neis1.remove(crul)
        mask2 = (df_link['source'] == None)
        for nei in neis1:
            mask2 |= ComputeNeighbor(nei, strong_relations)
        imp_neighbor2 = df_link[mask2]
        neis2 = list(set(imp_neighbor2['source'])) + list(set(imp_neighbor2['target']))
        for n1 in neis1:
            if n1 in neis2:
                neis2.remove(n1)
        mask3 = (df_link['source'] == None)
        imp_neighbor3 = df_link[mask3]
        neis3 = list(set(imp_neighbor3['source'])) + list(set(imp_neighbor3['target']))
        for n in neis3:
            mask3 |= ((df_link['source'] == n) | (df_link['target'] == n))
        mask3 &= ((df_link['relation']=='r_subdomain') |\
        (df_link['relation']=='r_cert')| (df_link['relation']=='r_request_jump') | (df_link['relation']=='r_dns_a'))
        mask |= (mask1 | mask2 | mask3)
    return mask
```
